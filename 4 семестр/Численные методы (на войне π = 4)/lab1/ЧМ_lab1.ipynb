# Задание 1

Найти такие n, при которых значение выражения $\frac{1 + 2^{-n} - 1}{2^{-n}}$ ≠ 1

### Теоретическое решение

Число с плавающей точкой в python:

| знак | экспонента (11 бит) | мантисса (52 бита) |
   
Число: $(-1)^{знак} * (1 + мантиса) * 2^{экспонента}$ 

Начиная с n=53 (т.к. в мантисе хранятся значения от $2^{-1}$ до $2^{-52}$)  $2^{-n}$ становится меньше младшего бита мантиссы числа 1, в следствие чего $1+2^{-n}=1$

```python
def num(n: int) -> float:
    epsilon = 2 ** -n
    return (1 + epsilon - 1) / epsilon

def main():
    print("n    дробь")
    for n in range(50, 56):
        print(n, " ", num(n))

if __name__ == "__main__":
    main()
```

# Задание 2

При больших n значение интеграла стремится к 0. 

float точно хранит значение до 15-16 цифр, т.е. n*rec(n-1) при может достигнуть значения чуть большего, чем 1.

1.00000... -  1.00011... = -0.00011...

Получается очень маленькое отрицитальное число, хотя должно быть очент маленькое положительное (18 шаг)

```python
import numpy as np
import math

def rec(n):
    if n == 1:
        return 1/np.exp(1)
    
    return 1 - n*rec(n-1)

mistake = 0.000000000000000001

print("  Интеграл               Ошибка")
for i in range(1, 31):
    print(f"{i} {rec(i):.17}   {math.factorial(i)*mistake:.17}")
```

# Задание 3
При сложении происходит выравнивание (приведение к одинаковой экспоненте). Число с меньшим порядком сдвигается вправо, при этом теряются младшие биты, что приводит к 

```python
import math

def factorial(x):
    if x == 1:
        return x
    return x*factorial(x-1)

space = lambda num : " " if num > 0 else ""

def sin_series(x, eps=1e-17, silent=True):
    _format=".5e"
    member = x  # k = 0: x^1 / 1!
    sin_x = member
    k = 1

    if not silent:
        print(f"{k} Член ряда: {space(member)}{format(member, _format)} Сумма ряда: {space(member)}{format(sin_x, _format)}")
        
    while abs((-1)**k * x**(2*k+1) / factorial(2*k+1)) > eps:
        member = (-1)**k * x**(2*k+1) / factorial(2*k+1)
        sin_x += member
        k += 1

        if not silent:
           print(f"{k} Член ряда: {space(member)}{format(member, _format)} Сумма ряда: {space(member)}{format(sin_x, _format)}")

    return sin_x

small_points_names = [0, "π/3", "π/2", "π", "2π"]
small_points = [0, math.pi/3, math.pi/2, math.pi, 2*math.pi]

large_points_names = ["12π", "13π", "14π"]
large_points = [12*math.pi, 13*math.pi, 14*math.pi]

eps = 1e-17

print("Сумма ряда при малом аргументе")
for i in range(len(small_points)):
    approx = sin_series(small_points[i], eps)
    print(f"x = {small_points_names[i]}: sin = {approx:.17f}\n")

print("\nСумма ряда при большом аргументе")
for i in range(len(large_points)):
    approx = sin_series(large_points[i], eps, silent=False)
    print(f"x = {large_points_names[i]}: sin = {approx:.16f}\n")
```
