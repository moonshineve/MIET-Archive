```python
import numpy as np
import matplotlib.pyplot as plt


def f(t):
    return np.exp(t)/3
    
x_experemental = np.linspace(0, 1, 5)
x_real = np.linspace(0, 1, 100)

plt.grid()
plt.axvline(x=0, color="gray")
plt.axhline(y=0, color="gray")

plt.plot(x_experemental, f(x_experemental), label="Интерполированная функция")
plt.plot(x_real, f(x_real), label="Исходная функция")
plt.scatter(x_experemental, f(x_experemental), color="black", label="Точки интерполяции")

plt.legend(loc = "upper left")
plt.title("Линейная интерполяция")
```
